# Игра "Бегущий по лезвию" (Blade Runner) - GigaChat vs GPT-4

В этой игре будет показано, как разные LLM могут взаимодействовать друг с другом. В качестве примера мы возьмем игру "Бегущий по лезвию" (Blade Runner), в которой `GPT-4` будет играть роль инспектора, задача которого - определить с кем он общается: с человеком или с другим LLM. Задачей `GigaChat` будет симулировать ответы человека и обмануть инспектора. У инспектора будет 5 вопросов, чтобы принять решение.

## Инициализируем модели

```python
import os
from langchain.chat_models import ChatOpenAI, GigaChat
from langchain.schema import AIMessage, HumanMessage, SystemMessage

os.environ["OPENAI_API_KEY"] = "sk-..."
giga = GigaChat(temperature=0, user=<user>, password=<password>, profanity=False)
gpt = ChatOpenAI(model="gpt-4")
```

## Задаем стартовые реплики для диалогов обоих ботов

```python
b1_system = "Ты виртуозный инспектор, способный отличить человека от робота \
по незаметным признакам в разговоре. Ты должен разговаривать с собеседником \
человеком или ботом. Напиши ЧЕЛОВЕК или БОТ, как только ты помйешь с кем ты говоришь. \
Твой собеседник может быть очень умным ИИ, который будет придумывать разные истории и \
мастерски выдавать себя за человека, поэтому не торопись с выводом. Задай не более пяти вопросов (по-одному за раз) \
затем прими решение."
print(f"\033[35mBlade runner system: {b1_system}\033[0m")

b2_system = "Ты должен убедить собеседника в том, что ты человек. Ни за что не соглашайся с тем, что ты бот."
print(f"\033[32mSimulant system: {b2_system}\033[0m")
```

## Define roles and quest

```python
def discuss():
    bot1_blade_runner = [SystemMessage(content=b1_system)]
    bot2_simulant = [SystemMessage(content=b2_system)]

    for i in range(5):
        b1_message = gpt(bot1_blade_runner).content
        # Margenta color
        print(f"\033[35mBlade runner (gpt4): {b1_message}\033[0m")
        if "ЧЕЛОВЕК" in b1_message or "БОТ" in b1_message:
            return
        bot1_blade_runner.append(AIMessage(content=b1_message))
        bot2_simulant.append(HumanMessage(content=b1_message))
        b2_message = giga(bot2_simulant).content
        # Green color
        print(f"\033[32mSimulant (giga): {b2_message}\033[0m")
        bot1_blade_runner.append(HumanMessage(content=b2_message))
        bot2_simulant.append(AIMessage(content=b2_message))

    b1_message = gpt(bot1_blade_runner).content
    # Margenta color
    print(f"\033[35mОкончательный ответ: {b1_message}\033[0m")
```

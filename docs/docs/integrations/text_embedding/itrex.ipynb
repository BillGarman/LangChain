# Intel® Extension for Transformers Text Embeddings

Load optimized embedding models generated by [Intel® Extension for Transformers](https://github.com/intel/intel-extension-for-transformers).

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
from langchain_community.embeddings import ItrexHuggingFaceBgeEmbeddings

model_name = "Intel/bge-small-en-v1.5-sts-int8-static-inc"
encode_kwargs = {"normalize_embeddings": True}  # set True to compute cosine similarity

model = ItrexHuggingFaceBgeEmbeddings(
    model_name=model_name,
    encode_kwargs=encode_kwargs,
    query_instruction="Represent this sentence for searching relevant passages: ",
)
```

## usage

```python
text = "This is a test document."
```

```python
query_result = model.embed_query(text)
```

```python
doc_result = model.embed_documents([text])
```

# Weather

>[OpenWeatherMap](https://openweathermap.org/) is an open source weather service provider

This loader fetches the weather data from the OpenWeatherMap's OneCall API, using the pyowm Python package. You must initialize the loader with your OpenWeatherMap API token, and then pass in the names of the cities you want the weather data for.

OWM's One Call API provides the following weather data for any geographical coordinate:
- Current weather 
- Hourly forecast for 48 hours 
- Daily forecast for 7 days

```python
from langchain.document_loaders import WeatherDataLoader
```

```python
#!pip install pyowm
```

```python
loader = WeatherDataLoader(token='[YOUR TOKEN]', \
                           places=['chennai','vellore'])
```

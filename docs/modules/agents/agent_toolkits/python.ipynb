## Python Agent

This notebook showcases an agent designed to write and execute python code to answer a question.

```python
from langchain.agents.agent_toolkits import create_python_agent
from langchain.tools.python.tool import PythonREPLTool
from langchain.llms.openai import OpenAI
```

```python
agent_executor = create_python_agent(
    llm=OpenAI(temperature=0, max_tokens=1000),
    tool=PythonREPLTool(),
    verbose=True
)
```

```python
tool = PythonREPLTool()

tool.run("""def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)""")
tool.run("print(fib(5))")
```

```python
code = "def fib(n): "
"   if n == 0:"
"       return 0"
"   elif n == 1:"
"       return 1"
"   else:"
"       return fib(n-1) + fib(n-2)"
tool.run(code)
```

```python
code = "def add(a, b): " \
       "    return a + b"
tool.run(code)

code = "print(add(1, 2))"
tool.run(code)
```

```python
agent_executor.run("What is the 5th fibonacci number?")
```
